```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file = "c:/Users/cvargas/Desktop/Pandas_Challenge/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file)

purchase_data.head()
```

```python
#Total Players
total_players = purchase_data["SN"].nunique()
total_players
```

```python
#Purchasing Analysis (Total)
##Number of Unique Items
unique_items = purchase_data["Item ID"].nunique() 
unique_items
```

```python
##Average Price
avg_price = purchase_data["Price"].mean()
avg_price
```

```python
##Total Number of Purchases
total_nbr_purch = len(purchase_data.index)
total_nbr_purch
```

```python
##Total Revenue
total_revenue = purchase_data["Price"].sum()
total_revenue
```

```python
#Gender Demographics
##Percentage and Count of Players by Gender
heroes_gender = pd.DataFrame(purchase_data.drop_duplicates(["SN"]))
gender_demo = heroes_gender.groupby(['Gender'])
gender = pd.DataFrame({"Total Count":gender_demo["Gender"].count(),"Percentage of players":gender_demo["Gender"].count()/total_players*100})
gender["Percentage of players"] = gender["Percentage of players"].map("{:.2f}%".format)
gender.head()
```

```python
#Purchasing analysis
purch = pd.DataFrame(purchase_data)
gender1 = purch.groupby(['Gender'])
gen_purch1 = pd.DataFrame({"Purchase Count":gender1["Price"].count(), "Average Purchase Price":gender1["Price"].mean(),"Total Purchase Value":gender1["Price"].sum(), "Normalized Totals":gender1["Price"].sum()/gender["Total Count"]})
gen_purch1["Average Purchase Price"] = gen_purch1["Average Purchase Price"].map("${:.2f}".format)
gen_purch1["Total Purchase Value"] = gen_purch1["Total Purchase Value"].map("${:,.2f}".format)
gen_purch1["Normalized Totals"] = gen_purch1["Normalized Totals"].map("${:,.2f}".format)
gen_purch1.head()
```

```python
age = pd.DataFrame(purchase_data)
group_names = ['<10', '10-13', '14-17', '18-21', '22-25', '26-29', '30-33', '34-37', '38-40', '>40']
years = [0, 9, 13, 17, 21, 25, 29, 33, 37, 40, 80]
age_demo = age.groupby(pd.cut(age["Age"], years, labels=group_names))
age2= pd.DataFrame({"Purchase Count":age_demo["Price"].count(), "Average Purchase Price":age_demo["Price"].mean(),"Total Purchase Value":age_demo["Price"].sum(),"Normalized Totals":age_demo["Price"].sum()/age_demo["SN"].nunique()})
age2["Average Purchase Price"] = age2["Average Purchase Price"].map("${:.2f}".format)
age2["Total Purchase Value"] = age2["Total Purchase Value"].map("${:,.2f}".format)
age2["Normalized Totals"] = age2["Normalized Totals"].map("${:,.2f}".format)
age2
```

```python
#Top Spenders
spenders = pd.DataFrame(purchase_data)
top_spend = spenders.groupby(['SN'])
top_spend2 = pd.DataFrame({"Purchase Count":top_spend["Price"].count(), "Average Purchase Price":top_spend["Price"].mean(),"Total Purchase Value":top_spend["Price"].sum()})
top_spend2 = top_spend2.sort_values("Total Purchase Value", ascending=False)
top_spend2["Average Purchase Price"] = top_spend2["Average Purchase Price"].map("${:.2f}".format)
top_spend2["Total Purchase Value"] = top_spend2["Total Purchase Value"].map("${:,.2f}".format)
top_spend2.head()
```

```python
profit_df = pd.DataFrame(purchase_data)
profit_grp = profit_df.groupby(['Item ID','Item Name'])
profit_df = pd.DataFrame({"Purchase Count":profit_grp["Price"].count(), "Item Price":profit_grp["Price"].mean(),"Total Purchase Value":profit_grp["Price"].sum()})
profit_df = profit_df.sort_values("Total Purchase Value", ascending=False)
profit_df["Item Price"] = profit_df["Item Price"].map("${:.2f}".format)
profit_df["Total Purchase Value"] = profit_df["Total Purchase Value"].map("${:,.2f}".format)
profit_df.head()
```
